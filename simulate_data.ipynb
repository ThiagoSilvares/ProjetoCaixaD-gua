```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Função para gerar dados fictícios com prefixos
def generate_data():
    data = []
    timestamps = pd.date_range(start="2024-01-01", periods=24, freq='H')  # 24 horas de dados
    for ts in timestamps:
        data.append({
            "name": "ArtesianWell",
            "fields": {
                "data_boardVoltage_ArtesianWell": np.random.uniform(1.5, 2.0),
                "data_waterLevel_ArtesianWell": np.random.uniform(1, 10),  # Nível de água do poço
                "rxInfo_rssi_0_ArtesianWell": np.random.randint(-100, -90),
                "rxInfo_loRaSNR_0_ArtesianWell": np.random.uniform(10, 15),
                "rxInfo_latitude_0_ArtesianWell": -23.64946,
                "rxInfo_longitude_0_ArtesianWell": -46.57367,
            },
            "timestamp": ts
        })
        data.append({
            "name": "Hidrometer",
            "fields": {
                "data_boardVoltage_Hidrometer": np.random.uniform(1.5, 2.0),
                "data_flowRate_Hidrometer": np.random.uniform(0.1, 5.0),  # Vazão do hidrometro
                "rxInfo_rssi_0_Hidrometer": np.random.randint(-100, -90),
                "rxInfo_loRaSNR_0_Hidrometer": np.random.uniform(10, 15),
                "rxInfo_latitude_0_Hidrometer": -23.64891,
                "rxInfo_longitude_0_Hidrometer": -46.57413,
            },
            "timestamp": ts
        })
        data.append({
            "name": "SmartLights",
            "fields": {
                "data_status_SmartLights": np.random.choice(["on", "off"]),  # Estado da luz
                "rxInfo_rssi_0_SmartLights": np.random.randint(-100, -90),
                "rxInfo_loRaSNR_0_SmartLights": np.random.uniform(10, 15),
                "rxInfo_latitude_0_SmartLights": -23.64946,
                "rxInfo_longitude_0_SmartLights": -46.57367,
            },
            "timestamp": ts
        })
        data.append({
            "name": "WaterTankLevel",
            "fields": {
                "data_boardVoltage_WaterTankLevel": np.random.uniform(0.1, 0.2),
                "data_waterLevel_WaterTankLevel": np.random.uniform(1, 10),  # Nível de água do tanque
                "rxInfo_rssi_0_WaterTankLevel": np.random.randint(-100, -90),
                "rxInfo_loRaSNR_0_WaterTankLevel": np.random.uniform(10, 15),
                "rxInfo_latitude_0_WaterTankLevel": -23.64946,
                "rxInfo_longitude_0_WaterTankLevel": -46.57367,
            },
            "timestamp": ts
        })
    return data

# Gerar dados fictícios
data = generate_data()

# Converter os dados para um DataFrame
df = pd.DataFrame([
    {
        "name": d["name"],
        "timestamp": d["timestamp"],
        **d["fields"]
    }
    for d in data
])

df["timestamp"] = pd.to_datetime(df["timestamp"])

# Exibir o DataFrame
df.head(48)  # Exibir as primeiras 48 linhas para ver os dados gerados
```

```python
plt.figure(figsize=(10, 6))
for column in ['data_boardVoltage_ArtesianWell', 'data_boardVoltage_Hidrometer', 'data_boardVoltage_WaterTankLevel']:
    plt.plot(df['timestamp'], df[column], marker='o', label=column.split('_')[2])

plt.xlabel('Timestamp')
plt.ylabel('Board Voltage')
plt.title('Board Voltage over Time')
plt.legend()
plt.grid(True)
plt.show()
```

RSSI (Received Signal Strength Indicator) é uma medida da potência do sinal recebido por um dispositivo em uma rede sem fio. Essa métrica é geralmente expressa em decibéis-miliwatts (dBm) e indica a força do sinal que um receptor (por exemplo, um sensor ou um roteador) recebe de um transmissor (por exemplo, um dispositivo IoT ou um ponto de acesso).
Unidade: dBm (decibéis em relação a 1 miliwatt).
Valores Típicos:

    0 dBm representa 1 miliwatt de potência de sinal.
    Valores negativos, como -50 dBm ou -100 dBm, são comuns, onde -50 dBm representa um sinal mais forte do que -100 dBm.

Interpretação:

    Valores mais próximos de 0 indicam um sinal mais forte.
    Valores mais baixos (mais negativos) indicam um sinal mais fraco.


Uso Prático do RSSI:

    Diagnóstico de Rede: Ajuda a diagnosticar problemas de cobertura e qualidade de sinal em redes sem fio.
    Localização de Dispositivos: Utilizado em técnicas de localização para estimar a distância entre dispositivos com base na força do sinal.
    Otimização de Conexão: Auxilia na escolha de pontos de acesso ou repetidores ideais para otimizar a cobertura de rede.

Exemplo de Interpretação dos Valores de RSSI:

    -30 dBm: Excelente sinal, muito forte.
    -50 dBm a -60 dBm: Bom sinal, adequado para a maioria das aplicações.
    -70 dBm a -80 dBm: Sinal fraco, pode haver problemas de conectividade.
    -90 dBm e abaixo: Sinal muito fraco, conexão instável ou impossível.

Contexto no Código:

No seu caso, rxInfo_rssi_0 é um campo que contém o valor de RSSI recebido pelo dispositivo. Ele ajuda a avaliar a qualidade da conexão do dispositivo com a rede, e gráficos comparando o RSSI entre diferentes dispositivos podem fornecer insights sobre a cobertura da rede e a localização dos dispositivos.

LoRaSNR (LoRa Signal-to-Noise Ratio) é uma métrica que indica a qualidade do sinal recebido em uma rede LoRa (Long Range). O SNR é a razão entre a potência do sinal recebido e o ruído de fundo, expressa em decibéis (dB). No contexto de redes LoRa, o SNR é um parâmetro crucial para determinar a robustez e a qualidade da comunicação sem fio.
    Unidade: Decibéis (dB).
    Valores Típicos:
        Valores positivos indicam que o sinal é mais forte do que o ruído de fundo.
        Valores negativos indicam que o sinal é mais fraco do que o ruído de fundo.
    Interpretação:
        Valores Altos (positivos): Indicam uma boa qualidade de sinal. Um SNR mais alto significa que o sinal é claramente distinguível do ruído.
        Valores Baixos (negativos): Indicam uma pior qualidade de sinal. Um SNR mais baixo significa que o sinal está próximo do nível de ruído, o que pode resultar em maior probabilidade de erros de comunicação.

Uso Prático do LoRaSNR:

    Diagnóstico de Rede: Ajuda a avaliar a qualidade geral da rede e a identificar áreas com interferências ou problemas de cobertura.
    Desempenho de Comunicação: SNR é utilizado para ajustar parâmetros da rede, como a taxa de dados e a potência de transmissão, para otimizar a comunicação.
    Análise de Robustez: Em ambientes ruidosos, um SNR alto é essencial para manter a integridade dos dados transmitidos.

Exemplo de Interpretação dos Valores de LoRaSNR:

    >10 dB: Excelente qualidade de sinal.
    6 dB a 10 dB: Boa qualidade de sinal.
    0 dB a 6 dB: Qualidade de sinal aceitável, mas pode ser suscetível a interferências.
    <0 dB: Qualidade de sinal ruim, o que pode resultar em alta taxa de erro.

Contexto no Código:

No seu caso, rxInfo_loRaSNR_0 é um campo que contém o valor de SNR para o sinal LoRa recebido pelo dispositivo. Este valor é importante para avaliar a qualidade da comunicação LoRa e pode ser utilizado para comparações entre dispositivos e para otimização da rede.

```python
import json
import matplotlib.pyplot as plt
import pandas as pd

# Carregar o arquivo JSON gerado
with open('data.json', 'r') as f:
    data = json.load(f)

# Criar listas para os dados de timestamps e water level de ArtesianWell
timestamps = []
water_level_data = []

# Extrair os dados do JSON
for entry in data:
    if entry['name'] == 'ArtesianWell':
        timestamp = pd.to_datetime(entry['timestamp'], unit='ns')
        water_level = entry['fields'].get('data_waterLevel')
        
        timestamps.append(timestamp)
        water_level_data.append(water_level)

# Convert to pandas Series to handle missing values and indexing
timestamps = pd.Series(timestamps)
water_level_data = pd.Series(water_level_data)

# Remover entradas onde os dados são None
water_level_data = water_level_data[water_level_data.notna()]

# Plotar o gráfico de water level (ArtesianWell)
plt.figure(figsize=(10, 5))
plt.plot(timestamps[:len(water_level_data)], water_level_data, label='Water Level (ArtesianWell)')
plt.xlabel('Time')
plt.ylabel('Water Level')
plt.title('Water Level Over Time for ArtesianWell')
plt.legend()
plt.grid(True)
plt.show()
```
